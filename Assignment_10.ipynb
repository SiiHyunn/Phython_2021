```python
import numpy as np
import math

t=np.arange(0,2.5,0.1)
y1=np.sin(math.pi*t)
y2=np.sin(math.pi*t+math.pi/2)
y3=np.sin(math.pi*t-math.pi/2)

plt.plot(t, y1, 'm*')
plt.plot(t, y2, 'g^')
plt.plot(t, y3, 'ys')

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(t, y1, 'm--')
plt.plot(t, y2, 'g')
plt.plot(t, y3, 'r-.')

plt.grid()
plt.legend()
plt.show()
```

```python
x=np.arange(-2*np.pi,2*np.pi,0.01)
y1=np.sin(3*x)/x
y2=np.sin(2*x)/x
y3=np.sin(x)/x

plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r')
plt.plot(x, y3, 'g')

plt.grid()
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
index=np.arange(5)
values1=[5,7,3,4,6]
values2=[6,6,4,5,7]
values3=[5,6,5,4,6]
bw=0.3

plt.title('A Multiseries Bar Chart')
plt.bar(index + 0.0, values1, color='b', width = bw)
plt.bar(index + 0.3, values2, color='g', width = bw)
plt.bar(index + 0.6, values3, color='r', width = bw)

ticklabel=['A','B','C','D','E']
plt.xticks(index, ticklabel, fontsize=12)
```

```python
series1=np.array([3,4,5,3])
series2=np.array([1,2,2,5])
series3=np.array([2,3,3,4])
index=np.arange(4)

bw=0.3

plt.title('A Multiseries Stacked Bar Chart')
plt.bar(index, series1, color='r', width = bw)
plt.bar(index, series2, color='b', width = bw, bottom=series1)
plt.bar(index, series3, color='g', width = bw, bottom=[sum(data) for data in zip(series1,series2)])
```

```python
plt.style.use('ggplot')

index=np.arange(8)
y1=np.array([1,3,4,6,4,3,2,1])
y2=np.array([1,2,5,4,3,3,2,1])

plt.bar(index, y1, color='r')
plt.bar(index, -y2, color='b')

for index, value in enumerate(y1):
    plt.text(index - 0.1, value + 0.5, str(value))
for index, value in enumerate(y2):
    plt.text(index - 0.1, -value - 0.5, str(value))
```

```python
from mpl_toolkits.mplot3d import Axes3D

plt.style.use('ggplot')
ax = plt.axes(projection ='3d')

def f(X,Y):
    return((1-Y**5+X**5)*np.exp(-X**2-Y**2))
    
X=np.arange(-2,2,0.1)
Y=np.arange(-2,2,0.1)
X,Y=np.meshgrid(X,Y)

ax.scatter3D(X,Y,f(X,Y))
```
